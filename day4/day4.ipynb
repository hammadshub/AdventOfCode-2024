**Problem Statement:**
Count all occurrences of the word "XMAS" in a given word search grid, where the word can appear horizontally, vertically, diagonally, forwards, or backwards, including overlapping instances.

```python
# Upload the word search file
from google.colab import files
uploaded = files.upload()

# Ensure the uploaded file is used
file_name = list(uploaded.keys())[0]  # Get the name of the uploaded file

# Open the uploaded file
with open(file_name, 'r') as file:
    word_search_data = file.read()

# Prepare the grid from the file content
word_search_lines = word_search_data.strip().split('\n')
grid = [list(line) for line in word_search_lines]

# Function to count "XMAS" occurrences in all directions
def count_xmas(grid):
    rows, cols = len(grid), len(grid[0])
    xmas_count = 0
    directions = [
        (0, 1),   # Right
        (1, 0),   # Down
        (1, 1),   # Diagonal down-right
        (1, -1),  # Diagonal down-left
        (0, -1),  # Left
        (-1, 0),  # Up
        (-1, -1), # Diagonal up-left
        (-1, 1)   # Diagonal up-right
    ]

    # Helper function to check a word in a specific direction
    def check_word(row, col, dr, dc):
        word = []
        for i in range(4):  # Length of "XMAS"
            r, c = row + i * dr, col + i * dc
            if 0 <= r < rows and 0 <= c < cols:
                word.append(grid[r][c])
            else:
                return False
        return ''.join(word) == "XMAS"

    # Traverse each cell in the grid
    for r in range(rows):
        for c in range(cols):
            for dr, dc in directions:
                if check_word(r, c, dr, dc):
                    xmas_count += 1

    return xmas_count

# Count occurrences of "XMAS"
total_xmas = count_xmas(grid)
print(f"The word 'XMAS' appears a total of {total_xmas} times.")
```

**Problem Statement:**
Count how many times the "X-MAS" pattern appears in a word search grid. The "X-MAS" pattern consists of two "MAS" arranged in the shape of an 'X', with 'M' and 'S' at the diagonals and 'A' at the center. The pattern can appear both forwards and backwards.

```python
# Upload the word search file
from google.colab import files
uploaded = files.upload()

# Read the uploaded file
file_name = list(uploaded.keys())[0]  # Get the name of the uploaded file

# Function to count X-MAS patterns
def count_xmas_patterns(grid):
    rows, cols = len(grid), len(grid[0])
    xmas_count = 0

    # Check for X-MAS patterns centered at each valid cell
    for r in range(1, rows - 1):  # Avoid the grid edges
        for c in range(1, cols - 1):  # Avoid the grid edges
            if grid[r][c] == 'A':  # Center of the X-MAS must be 'A'
                # Extract the four corners of the X-MAS
                top_left = grid[r - 1][c - 1]
                top_right = grid[r - 1][c + 1]
                bottom_left = grid[r + 1][c - 1]
                bottom_right = grid[r + 1][c + 1]

                # Check if diagonals form valid MAS or SAM (forwards or backwards)
                if ((top_left == 'M' and bottom_right == 'S') or (top_left == 'S' and bottom_right == 'M')) and \
                   ((top_right == 'M' and bottom_left == 'S') or (top_right == 'S' and bottom_left == 'M')):
                    xmas_count += 1

    return xmas_count

# Read the content of the uploaded file
with open(file_name, 'r') as file:
    word_search_data = file.read()

# Prepare the grid from the file content
word_search_lines = word_search_data.strip().split('\n')
grid = [list(line) for line in word_search_lines]

# Count occurrences of X-MAS patterns in the grid
xmas_patterns_total = count_xmas_patterns(grid)
print(f"The word 'X-MAS' appears a total of {xmas_patterns_total} times.")
```
